```python
%matplotlib inline
import nltk
import string

```

```python
sent = "Awe very properly hangs about it, since it is the immovable standard and silent witness " 
sent += "of all our memories and assertions; and the past and the future, which in our anxious life "
sent += "are so differently interesting and so differently dark, are one seamless garment for the truth, "
sent += "shining like the sun."

print(sent)
```

```python
tokens = nltk.word_tokenize(sent)
tagged_tokens = nltk.pos_tag(tokens)
import json
print(json.dumps(tagged_tokens, indent=2))
```

```python
nltk.download('all')
sent_detector = nltk.data.load('tokenizers/punkt/english.pickle')
```

```python
text = open('guten-0-0-1.txt').read()
lines = text.split('\n')
count = 0
lineno = 0
for lineno, line in enumerate(lines):
    if '===========================================' in line:
        count += 1
    if count >= 5:
        break
lineno += 1
text = '\n'.join(lines[lineno:])
with open('preamble.txt', 'w') as fout:
    fout.write(text)
    

print(text)

```

```python
text = open('preamble.txt').read()
sents = sent_detector.tokenize(text.strip())

for i, sent in enumerate(sents):
    # print('-'*80)
    sents[i] = sent.replace('\r', ' ').replace('\n', ' ')
    print('{:05d}: {}'.format(i, sent))

sents = sents[:100]
```

```python
for i, sent in enumerate(sents):
    sent = sent.lower()
    sent = sent.translate(None, string.punctuation)
    tokens = nltk.word_tokenize(sent)
    tagged_tokens = nltk.pos_tag(tokens)
    print('-'*80)
    print('{:03d}: {}'.format(i, tagged_tokens))
    # print(tagged_tokens)
```
