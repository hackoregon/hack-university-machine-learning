### Your First DataFrame
Load some crime data into a dictionary of pandas data frames

```python
import pandas as pd
crime = {}
for yr in range(13, 15):
    crime[yr] = pd.DataFrame.from_csv('crime_incident_data_20{}.csv'.format(yr))
df = crime[13]
df
```

### Explore the Data
`df.describe()` works well for numerical data, but `pandas_profiling.describe(df)` works for text and categorical data

```python
df.describe()
```

```python
import pandas_profiling as pdp
stats = pdp.describe(df)
stats
```

```python
df = crime[13].copy()
labels = list(df.columns)
for i in range(-1, -3, -1):
    labels[i] = labels[i].strip().lower()[0]
df = pd.DataFrame(df.values, columns=labels)
plt.scatter(df.x, df.y)
```
